```python
import math as m
import swisseph as swe
from datetime import datetime
import import_ipynb
import func as f
from termcolor import colored
from dateutil.relativedelta import relativedelta
from IPython.display import display, Markdown, Latex

#display(Markdown('*some markdown* $\phi$'))
# If you particularly want to display maths, this is more direct:
#display(Latex('\phi'))

##########################################################################################
# User Parameter dob = '23/08/1963 7:29:00' Date of Birth
##########################################################################################
dob = '23/08/1963 7:29:00'
cDate = '10/01/2015 13:56:00'
tZone = 5.5 # time Zone
# latitude = 24.50
# longitude = 93.5
##########################################################################################
try:
    dBirth = datetime.strptime(dob, '%d/%m/%Y %H:%M:%S')
    dCurrent = datetime.strptime(cDate, '%d/%m/%Y %H:%M:%S')
except ValueError as ve:
    print('ValueError Raised:', ve)

fDOBHour = (dBirth.hour + (dBirth.minute/60) + (dBirth.second/3600)) # / 24 - 23.9344444444444444
fDOBHourUTC = fDOBHour - tZone
jBirthDateTime = swe.date_conversion(dBirth.year,dBirth.month,dBirth.day,fDOBHourUTC)

fCDateHour = (dCurrent.hour + (dCurrent.minute/60) + (dCurrent.second/3600))
fCDateHourUTC = fCDateHour - tZone
jCurrentDateTime = swe.date_conversion(dCurrent.year,dCurrent.month,dCurrent.day,fCDateHourUTC)

r = swe.set_sid_mode(1,0,0)
lunarLongitude = swe.calc_ut(jBirthDateTime[1],1,65536)
nakshaID = f.nakshatraID(lunarLongitude[0][0])

lunarSign = lunarLongitude[0][0]/30
lunarDegree = m.modf(lunarSign)[0]*30
lunarMinute = m.modf(lunarDegree)[0]*60
lunarSecond = m.modf(lunarMinute)[0]*60

##########################################################################################
# Display Result Lunar Position
##########################################################################################
display(Markdown('<font color="blue"><h3>Lunar Postion</h3></font>'))
print("Lunar Sign: {0} Degree: {1} Minute: {2} Second: {3:5.2f}".format(\
                                            m.floor(lunarSign),\
                                            m.floor(lunarDegree),\
                                            m.floor(lunarMinute),\
                                            lunarSecond))
##########################################################################################
# Astottari Dasha Calculation based on Kritikadi System
##########################################################################################
# Astottari Bhukta
display(Markdown('<font color="blue"><h3>Dasha Calculation</h3></font>'))
astoGraha = f.astottariDashaGraha(nakshaID)
astoBhukta = f.astottariDasha(lunarLongitude[0][0],1)
astoBhuktaYear = astoBhukta
astoBhuktaMonth = m.modf(astoBhuktaYear)[0]*12
astoBhuktaDay = m.modf(astoBhuktaMonth)[0]*30
astoBhuktaDand = m.modf(astoBhuktaDay)[0]*60
# if m.ceil(astoBhuktaDand) == 60:
#     astoBhuktaDay += 1
#     astoBhuktaDand = 0
#     if m.floor(astoBhuktaDay) == 30:
#         astoBhuktaMonth += 1
#         astoBhuktaDay = 0
#         if m.floor(astoBhuktaMonth) == 12:
#             astoBhuktaYear +=1
#             astoBhuktaMonth = 0

# Astottari Bhogya
astoBhogya = f.astottariDasha(lunarLongitude[0][0],2)
astoBhogyaYear = astoBhogya
astoBhogyaMonth = m.modf(astoBhogyaYear)[0]*12
astoBhogyaDay = m.modf(astoBhogyaMonth)[0]*30
astoBhogyaDand = m.modf(astoBhogyaDay)[0]*60
# Display result
print(colored("\nAstottari Dasha calculated in Kritikadi System:",'blue'))
print("Bhukta Dasha ({0}):  Year: {1:2d} Month: {2:2d} Day: {3:2d} Danda: {4:5.2f}".format(\
                                            astoGraha, \
                                            m.floor(astoBhukta),\
                                            m.floor(astoBhuktaMonth),\
                                            m.floor(astoBhuktaDay),\
                                            astoBhuktaDand))
print("Bhogya Dasha ({0}):  Year: {1:2d} Month: {2:2d} Day: {3:2d} Danda: {4:5.2f}".format(\
                                            astoGraha, \
                                            m.floor(astoBhogya),\
                                            m.floor(astoBhogyaMonth),\
                                            m.floor(astoBhogyaDay),\
                                            astoBhogyaDand))

##########################################################################################
# Vimsottari Dasha Calculation
##########################################################################################
vimGraha = f.vimsottariDashaGraha(nakshaID)            
# Vimsottari Bhukta
vimBhukta = f.vimsottariDasha(lunarLongitude[0][0],1)
vimBhuktaYear = vimBhukta
vimBhuktaMonth = m.modf(vimBhuktaYear)[0]*12
vimBhuktaDay = m.modf(vimBhuktaMonth)[0]*30
vimBhuktaDand = m.modf(vimBhuktaDay)[0]*60
# if m.ceil(vimBhuktaDand) == 60:
#     vimBhuktaDay += 1
#     vimBhuktaDand = 0
#     if m.floor(vimBhuktaDay) == 30:
#         vimBhuktaMonth += 1
#         vimBhuktaDay = 0
#         if m.floor(vimBhuktaMonth) == 12:
#             vimBhuktaYear +=1
#             vimBhuktaMonth = 0

# Vimsottari Bhogya
vimBhogya = f.vimsottariDasha(lunarLongitude[0][0],2)
vimBhogyaYear = vimBhogya
vimBhogyaMonth = m.modf(vimBhogyaYear)[0]*12
vimBhogyaDay = m.modf(vimBhogyaMonth)[0]*30
vimBhogyaDand = m.modf(vimBhogyaDay)[0]*60

# Display result
print(colored("\nVimsottari Dasha",'blue'))
print("Bhukta Dasha ({0}):  Year: {1:2d} Month: {2:2d} Day: {3:2d} Danda: {4:5.2f}".format(\
                                            vimGraha, \
                                            m.floor(vimBhukta),\
                                            m.floor(vimBhuktaMonth),\
                                            m.floor(vimBhuktaDay),\
                                            vimBhuktaDand))
print("Bhogya Dasha ({0}):  Year: {1:2d} Month: {2:2d} Day: {3:2d} Danda: {4:5.2f}".format(\
                                            vimGraha, \
                                            m.floor(vimBhogya),\
                                            m.floor(vimBhogyaMonth),\
                                            m.floor(vimBhogyaDay),\
                                            vimBhogyaDand))

##########################################################################################
# Yogini Dasha Calculation
##########################################################################################

yogiGraha = f.yoginiDashaGraha(nakshaID)            
# Yogini Bhukta
yogiBhukta = f.yoginiDasha(lunarLongitude[0][0],1)
yogiBhuktaYear = yogiBhukta
yogiBhuktaMonth = m.modf(yogiBhuktaYear)[0]*12
yogiBhuktaDay = m.modf(yogiBhuktaMonth)[0]*30
yogiBhuktaDand = m.modf(yogiBhuktaDay)[0]*60
# if m.ceil(yogiBhuktaDand) == 60:
#     yogiBhuktaDay += 1
#     yogiBhuktaDand = 0
#     if m.floor(yogiBhuktaDay) == 30:
#         yogiBhuktaMonth += 1
#         yogiBhuktaDay = 0
#         if m.floor(astoBhuktaMonth) == 12:
#             yogiBhuktaYear +=1
#             yogiBhuktaMonth = 0

# Yogini Bhogya
yogiBhogya = f.yoginiDasha(lunarLongitude[0][0],2)
yogiBhogyaYear = yogiBhogya
yogiBhogyaMonth = m.modf(yogiBhogyaYear)[0]*12
yogiBhogyaDay = m.modf(yogiBhogyaMonth)[0]*30
yogiBhogyaDand = m.modf(yogiBhogyaDay)[0]*60


# Display result
print(colored("\nYogini Dasha",'blue'))
print("Bhukta Dasha ({0}):  Year: {1:2d} Month: {2:2d} Day: {3:2d} Danda: {4:5.2f}".format(\
                                            yogiGraha, \
                                            m.floor(yogiBhukta),\
                                            m.floor(yogiBhuktaMonth),\
                                            m.floor(yogiBhuktaDay),\
                                            yogiBhuktaDand))
print("Bhogya Dasha ({0}):  Year: {1:2d} Month: {2:2d} Day: {3:2d} Danda: {4:5.2f}".format(\
                                            yogiGraha, \
                                            m.floor(yogiBhogya),\
                                            m.floor(yogiBhogyaMonth),\
                                            m.floor(yogiBhogyaDay),\
                                            yogiBhogyaDand))

##########################################################################################
# Astottari Maha Dasha Calculation
##########################################################################################
display(Markdown('<font color="blue"><h3>Maha Dasha</h3></font>'))
# Astottari Maha Bhukta
astoMahaBhukta = f.astottariMahaDasha(jBirthDateTime[1], jCurrentDateTime[1], lunarLongitude, \
                   nakshaID, astoBhukta, astoGraha, 2)
astoMahaBhuktaYear = astoMahaBhukta
astoMahaBhuktaMonth = m.modf(astoMahaBhuktaYear)[0]*12
astoMahaBhuktaDay = m.modf(astoMahaBhuktaMonth)[0]*30
astoMahaBhuktaDand = m.modf(astoMahaBhuktaDay)[0]*60

# Astottari Maha Bhogya
astoMahaBhogya = f.astottariMahaDasha(jBirthDateTime[1], jCurrentDateTime[1], lunarLongitude, \
                   nakshaID, astoBhukta, astoGraha, 3)
astoMahaBhogyaYear = astoMahaBhogya
astoMahaBhogyaMonth = m.modf(astoMahaBhogyaYear)[0]*12
astoMahaBhogyaDay = m.modf(astoMahaBhogyaMonth)[0]*30
astoMahaBhogyaDand = m.modf(astoMahaBhogyaDay)[0]*60

astoMahaGraha = f.astottariMahaDasha(jBirthDateTime[1], jCurrentDateTime[1], lunarLongitude, \
                   nakshaID, astoBhukta, astoGraha, 4)

astoMahaEndTimeDays = (astoMahaBhogya * 360)
astoMahaEndTime = dCurrent + relativedelta(days=+astoMahaEndTimeDays)

# Display result
print(colored("\nAstottari Maha Dasha",'blue'))
print("Bhukta Dasha ({0}):  Year: {1:2d} Month: {2:2d} Day: {3:2d} Danda: {4:5.2f}".format(\
                                            astoMahaGraha, \
                                            m.floor(astoMahaBhuktaYear),\
                                            m.floor(astoMahaBhuktaMonth),\
                                            m.floor(astoMahaBhuktaDay),\
                                            astoMahaBhuktaDand))
print("Bhogya Dasha ({0}):  Year: {1:2d} Month: {2:2d} Day: {3:2d} Danda: {4:5.2f}".format(\
                                            astoMahaGraha, \
                                            m.floor(astoMahaBhogyaYear),\
                                            m.floor(astoMahaBhogyaMonth),\
                                            m.floor(astoMahaBhogyaDay),\
                                            astoMahaBhogyaDand))
print("Astottari Maha Dasha ({0}) ending at  Year: {1}.".format(\
                                            astoMahaGraha, \
                                            astoMahaEndTime))

##########################################################################################
# Astottari Antar Dasha Calculation
##########################################################################################
astoMahaID = f.astottariMahaDasha(jBirthDateTime[1], jCurrentDateTime[1], lunarLongitude, \
                   nakshaID, astoBhukta, astoGraha, 1)
astoAntarID = f.astottariAntarDasha(astoMahaID, astoMahaBhukta, 1)
astoAntarGraha = f.astottariAntarDasha(astoMahaID, astoMahaBhukta, 4)
astoAntarBhukta = f.astottariAntarDasha(astoMahaID, astoMahaBhukta, 2)
astoAntarBhuktaYear = astoAntarBhukta
astoAntarBhuktaMonth = m.modf(astoAntarBhuktaYear)[0]*12
astoAntarBhuktaDay = m.modf(astoAntarBhuktaMonth)[0]*30
astoAntarBhuktaDand = m.modf(astoAntarBhuktaDay)[0]*60

astoAntarBhogya = f.astottariAntarDasha(astoMahaID, astoMahaBhukta, 3)
astoAntarBhogyaYear = astoAntarBhogya
astoAntarBhogyaMonth = m.modf(astoAntarBhogyaYear)[0]*12
astoAntarBhogyaDay = m.modf(astoAntarBhogyaMonth)[0]*30
astoAntarBhogyaDand = m.modf(astoAntarBhogyaDay)[0]*60

# Display result
print(colored("\nAstottari Antar Dasha",'blue'))
print("Bhukta Dasha ({0}):  Year: {1:2d} Month: {2:2d} Day: {3:2d} Danda: {4:5.2f}".format(\
                                            astoAntarGraha, \
                                            m.floor(astoAntarBhuktaYear),\
                                            m.floor(astoAntarBhuktaMonth),\
                                            m.floor(astoAntarBhuktaDay),\
                                            astoAntarBhuktaDand))
print("Bhogya Dasha ({0}):  Year: {1:2d} Month: {2:2d} Day: {3:2d} Danda: {4:5.2f}".format(\
                                            astoAntarGraha, \
                                            m.floor(astoAntarBhogyaYear),\
                                            m.floor(astoAntarBhogyaMonth),\
                                            m.floor(astoAntarBhogyaDay),\
                                            astoAntarBhogyaDand))
   
##########################################################################################
# Vimsottari Maha Dasha Calculation
##########################################################################################    
vimMahaID = f.vimsottariMahaDasha(jBirthDateTime[1], jCurrentDateTime[1], lunarLongitude,nakshaID,vimBhukta, 1)
vimMahaGraha = f.vimsottariMahaDasha(jBirthDateTime[1], jCurrentDateTime[1], lunarLongitude,nakshaID,vimBhukta, 4)
vimMahaBhukta = f.vimsottariMahaDasha(jBirthDateTime[1], jCurrentDateTime[1], lunarLongitude,nakshaID,vimBhukta, 2)
vimMahaBhuktaYear = vimMahaBhukta
vimMahaBhuktaMonth = m.modf(vimMahaBhuktaYear)[0]*12
vimMahaBhuktaDay = m.modf(vimMahaBhuktaMonth)[0]*30
vimMahaBhuktaDand = m.modf(vimMahaBhuktaDay)[0]*60

vimMahaBhogya = f.vimsottariMahaDasha(jBirthDateTime[1], jCurrentDateTime[1], lunarLongitude,nakshaID,vimBhukta, 3)
vimMahaBhogyaYear = vimMahaBhogya
vimMahaBhogyaMonth = m.modf(vimMahaBhogyaYear)[0]*12
vimMahaBhogyaDay = m.modf(vimMahaBhogyaMonth)[0]*30
vimMahaBhogyaDand = m.modf(vimMahaBhogyaDay)[0]*60

vimMahaEndTimeDays = (vimMahaBhogya * 360)
vimMahaEndTime = dCurrent + relativedelta(days=+vimMahaEndTimeDays)

print(colored("\nVimsottari Maha Dasha",'blue'))
print("Bhukta Dasha ({0}):  Year: {1:2d} Month: {2:2d} Day: {3:2d} Danda: {4:5.2f}".format(\
                                            vimMahaGraha, \
                                            m.floor(vimMahaBhuktaYear),\
                                            m.floor(vimMahaBhuktaMonth),\
                                            m.floor(vimMahaBhuktaDay),\
                                            vimMahaBhuktaDand))
print("Bhogya Dasha ({0}):  Year: {1:2d} Month: {2:2d} Day: {3:2d} Danda: {4:5.2f}".format(\
                                            vimMahaGraha, \
                                            m.floor(vimMahaBhogyaYear),\
                                            m.floor(vimMahaBhogyaMonth),\
                                            m.floor(vimMahaBhogyaDay),\
                                            vimMahaBhogyaDand))
print("Vimsottari Maha Dasha ({0}) ending at  Year: {1}.".format(\
                                            vimMahaGraha, \
                                            vimMahaEndTime))

##########################################################################################
# Vimsottari Antar Dasha Calculation
##########################################################################################
vimAntarID = f.vimsottariAntarDasha(vimMahaID, vimMahaBhukta, 1)
vimAntarGraha = f.vimsottariAntarDasha(vimMahaID, vimMahaBhukta, 4)
vimAntarBhukta = f.vimsottariAntarDasha(vimMahaID, vimMahaBhukta, 2)
vimAntarBhuktaYear = vimAntarBhukta
vimAntarBhuktaMonth = m.modf(vimAntarBhuktaYear)[0]*12
vimAntarBhuktaDay = m.modf(vimAntarBhuktaMonth)[0]*30
vimAntarBhuktaDand = m.modf(vimAntarBhuktaDay)[0]*60

vimAntarBhogya = f.vimsottariAntarDasha(vimMahaID, vimMahaBhukta, 3)
vimAntarBhogyaYear = vimAntarBhogya
vimAntarBhogyaMonth = m.modf(vimAntarBhogyaYear)[0]*12
vimAntarBhogyaDay = m.modf(vimAntarBhogyaMonth)[0]*30
vimAntarBhogyaDand = m.modf(vimAntarBhogyaDay)[0]*60

# Display result
print(colored("\nVimsottari Antar Dasha",'blue'))
print("Bhukta Dasha ({0}):  Year: {1:2d} Month: {2:2d} Day: {3:2d} Danda: {4:5.2f}".format(\
                                            vimAntarGraha, \
                                            m.floor(vimAntarBhuktaYear),\
                                            m.floor(vimAntarBhuktaMonth),\
                                            m.floor(vimAntarBhuktaDay),\
                                            vimAntarBhuktaDand))
print("Bhogya Dasha ({0}):  Year: {1:2d} Month: {2:2d} Day: {3:2d} Danda: {4:5.2f}".format(\
                                            vimAntarGraha, \
                                            m.floor(vimAntarBhogyaYear),\
                                            m.floor(vimAntarBhogyaMonth),\
                                            m.floor(vimAntarBhogyaDay),\
                                            vimAntarBhogyaDand))
##########################################################################################
# Yogini Maha Dasha Calculation
##########################################################################################
yogiMahaID = f.yoginiMahaDasha(jBirthDateTime[1], jCurrentDateTime[1], lunarLongitude,nakshaID,yogiBhukta, 1)
yogiMahaGraha = f.yoginiMahaDasha(jBirthDateTime[1], jCurrentDateTime[1], lunarLongitude,nakshaID,yogiBhukta, 4)
yogiMahaBhukta = f.yoginiMahaDasha(jBirthDateTime[1], jCurrentDateTime[1], lunarLongitude,nakshaID,yogiBhukta, 2)
yogiMahaBhuktaYear = yogiMahaBhukta
yogiMahaBhuktaMonth = m.modf(yogiMahaBhuktaYear)[0]*12
yogiMahaBhuktaDay = m.modf(yogiMahaBhuktaMonth)[0]*30
yogiMahaBhuktaDand = m.modf(yogiMahaBhuktaDay)[0]*60

yogiMahaBhogya = f.yoginiMahaDasha(jBirthDateTime[1], jCurrentDateTime[1], lunarLongitude,nakshaID,yogiBhukta, 3)
yogiMahaBhogyaYear = yogiMahaBhogya
yogiMahaBhogyaMonth = m.modf(yogiMahaBhogyaYear)[0]*12
yogiMahaBhogyaDay = m.modf(yogiMahaBhogyaMonth)[0]*30
yogiMahaBhogyaDand = m.modf(yogiMahaBhogyaDay)[0]*60

yogiMahaEndTimeDays = (yogiMahaBhogya * 360)
yogiMahaEndTime = dCurrent + relativedelta(days=+yogiMahaEndTimeDays)

print(colored("\nYogini Maha Dasha",'blue'))
print("Bhukta Dasha ({0}):  Year: {1:2d} Month: {2:2d} Day: {3:2d} Danda: {4:5.2f}".format(\
                                            yogiMahaGraha, \
                                            m.floor(yogiMahaBhuktaYear),\
                                            m.floor(yogiMahaBhuktaMonth),\
                                            m.floor(yogiMahaBhuktaDay),\
                                            yogiMahaBhuktaDand))
print("Bhogya Dasha ({0}):  Year: {1:2d} Month: {2:2d} Day: {3:2d} Danda: {4:5.2f}".format(\
                                            yogiMahaGraha, \
                                            m.floor(yogiMahaBhogyaYear),\
                                            m.floor(yogiMahaBhogyaMonth),\
                                            m.floor(yogiMahaBhogyaDay),\
                                            yogiMahaBhogyaDand))
print("Yogini Maha Dasha ({0}) ending at  Year: {1}.".format(\
                                            yogiMahaGraha, \
                                            yogiMahaEndTime))

##########################################################################################
# Yogini Antar Dasha Calculation
##########################################################################################
def yoginiAntarDasha(dashaID, bhukta, option):
    gName = ["Ulka", "Siddha", "Sangkata", "Mangala", "Pingla", "Dhanya", "Bhramari", "Bhadrika"]
    yoginiSpan = [6,7,8,1,2,3,4,5]
    flag = 0
    index = dashaID % 8
    mahaDasha = dashaID % 8
    currentAge = bhukta
    while currentAge > 0:
        if flag == 0: 
            bhogya = (yoginiSpan[mahaDasha] * yoginiSpan[index] / 36) - currentAge
            currentAge = currentAge - (yoginiSpan[mahaDasha] * yoginiSpan[index] / 36)
            dashaID = index
            index = (index + 1) % 8
            flag = 1
        else:
            bhukta = currentAge
            bhogya = (yoginiSpan[mahaDasha] * yoginiSpan[index] / 36) - currentAge
            currentAge = currentAge - (yoginiSpan[mahaDasha] * yoginiSpan[index] / 36)
            dashaID = index
            index = (index + 1) % 8
    if option == 1:
        return dashaID
    elif option == 2:
        return bhukta
    elif option == 3:
        return bhogya
    elif option == 4:
        return gName[dashaID]

yogiAntarID = yoginiAntarDasha(yogiMahaID, yogiMahaBhukta, 1)
yogiAntarGraha = yoginiAntarDasha(yogiMahaID, yogiMahaBhukta, 4)
yogiAntarBhukta = yoginiAntarDasha(yogiMahaID, yogiMahaBhukta, 2)
yogiAntarBhuktaYear = yogiAntarBhukta
yogiAntarBhuktaMonth = m.modf(yogiAntarBhuktaYear)[0]*12
yogiAntarBhuktaDay = m.modf(yogiAntarBhuktaMonth)[0]*30
yogiAntarBhuktaDand = m.modf(yogiAntarBhuktaDay)[0]*60

yogiAntarBhogya = yoginiAntarDasha(yogiMahaID, yogiMahaBhukta, 3)
yogiAntarBhogyaYear = yogiAntarBhogya
yogiAntarBhogyaMonth = m.modf(yogiAntarBhogyaYear)[0]*12
yogiAntarBhogyaDay = m.modf(yogiAntarBhogyaMonth)[0]*30
yogiAntarBhogyaDand = m.modf(yogiAntarBhogyaDay)[0]*60

# Display result
print(colored("\nYogini Antar Dasha",'blue'))
print("Bhukta Dasha ({0}):  Year: {1:2d} Month: {2:2d} Day: {3:2d} Danda: {4:5.2f}".format(\
                                            yogiAntarGraha, \
                                            m.floor(yogiAntarBhuktaYear),\
                                            m.floor(yogiAntarBhuktaMonth),\
                                            m.floor(yogiAntarBhuktaDay),\
                                            yogiAntarBhuktaDand))
print("Bhogya Dasha ({0}):  Year: {1:2d} Month: {2:2d} Day: {3:2d} Danda: {4:5.2f}".format(\
                                            yogiAntarGraha, \
                                            m.floor(yogiAntarBhogyaYear),\
                                            m.floor(yogiAntarBhogyaMonth),\
                                            m.floor(yogiAntarBhogyaDay),\
                                            yogiAntarBhogyaDand))
    
```
